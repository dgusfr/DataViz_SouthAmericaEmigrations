```python
import pandas as pd
```

```python
df = pd.read_csv('/content/dataset.csv')
```

```python
##Analise de Tendencia de Emigração do Brasil ao longo dos anos
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]

brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

Criando gráfico de barras com .bar e alterando o tamanho da legenda do eixo x com set_tick_params

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12, 5))
ax.bar(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.set_xlabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=10)

plt.show()
```

### Ordenando o gráfico decrescentemente de barras horizontal:

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)

cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
cores = []
for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

plt.show()
```

###Adicionando anotações na Barra e retirando o eixo x

```python
cores = []
for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

# v = corresponde ao valor das barras, que dps são convertidas em string
for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
#retirar marcações nos eixos:
ax.tick_params(axis='both', which='both', length=0)

plt.show()
```

```python
sns.set_theme()

top_10 = df.sort_values('Total', ascending=False).head(10)
```

```python
ax =sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set(title="Países com maior imigração para o Canadá\n1980 a 2013",
       xlabel='Número de Imigrantes',
        ylabel='')
```

```python
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set(title="Países com maior imigração para o Canadá\n1980 a 2013",
       xlabel='Número de Imigrantes',
        ylabel='')

plt.show()
```

### Criação de Funções e alteraação de cores das paletas

```python
def gerar_gráfico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)
    ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
    ax.set_xlabel('Número de imigrantes', fontsize=14)
    ax.set_ylabel('')
    plt.show()
```

#### Gerando degrade de paleta de cores pelo Seaborn
##### 'r' de reverso, para ser a cor mais escura no maior valor numerico

```python
gerar_gráfico_paleta('Blues_r')
```

```python
gerar_gráfico_paleta('rocket')
```

```python
gerar_gráfico_paleta('coolwarm')
```

#### paleta de cores categoricas:

```python
gerar_gráfico_paleta("tab10")
```

```python
sns.set_theme(style='dark')
gerar_gráfico_paleta('tab10')
```

```python
sns.set_theme(style='whitegrid')
gerar_gráfico_paleta('tab10')
```

#### ticks ou traços nos valores

```python
sns.set_theme(style='ticks')
gerar_gráfico_paleta('tab10')
```

#Crianção de Gráfico interativos com Plotly Express

```python
import plotly.express as px
```

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes')
fig.show()
```

####Adicionando titulo e mudando layout

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes',
              title='Imigração dos Brasil para o Canadá no período de 1980 a 2013')

fig.update_layout(width=1000, height=500,
                  xaxis={'tickangle': -45},
                  font_family='Arial',
                  font_size=14,
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=22,
                  xaxis_title='Ano',
                  yaxis_title='Número de imigrantes')

fig.show()
```

###Alterando as cores com Plotly

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes',
              title='Imigração dos Brasil para o Canadá no período de 1980 a 2013')

fig.update_traces(line_color='green', line_width=4)

fig.update_layout(width=1000, height=500,
                  xaxis={'tickangle': -45},
                  font_family='Arial',
                  font_size=14,
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=22,
                  xaxis_title='Ano',
                  yaxis_title='Número de imigrantes')

fig.show()
```

é possivel criar o gráfico estático e excluir a visualização de um pais em especifico clicando em cima do nome dele

```python
fig = px.line(america_sul_final, x=america_sul_final.index, y=america_sul_final.columns, color='País', markers=True,
              title='Imigração dos países da América do Sul para o Canadá de 1980 a 2013')
fig.update_layout(
    xaxis={'tickangle': -45},
    xaxis_title='Ano',
    yaxis_title='Número de imigrantes')

fig.show()
```
